```
import pandas as pd

DATA_URL = 'http://128.138.93.164/aprd6342/data/TravelPonyFacebook.csv'
df = pd.read_csv(DATA_URL)
df

```

```
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['day_of_week'] = df['Start Date'].dt.day_name()
df
```

```
df['cost_per_impression'] = df['Amount Spent (USD)']/df['Impressions']
num_day = df.groupby('day_of_week').count()
n_d = num_day.loc[:, 'Start Date']



```

```
sum_day = df.groupby('day_of_week').sum()
s_d = sum_day.loc[:, 'cost_per_impression']
```

```
ndf = pd.concat([n_d,s_d],axis=1,join_axes=[n_d.index])
ndf

ndf['avg_impression'] = ndf['cost_per_impression']/ndf['Start Date']   #sum all cost per impression/ numbers of days
ndf

ndf.sort_values(by=['avg_impression'],ascending=False)
```

CONCLUSION:

Frm the analytics result, we knows that "Firday" has the highest cost per impression and "Saturday" has the lowest cost per impression. That is to say,it cost most on Friday to attract customers and on Saturday it cost less.

```
lrModel = LinearRegression()

x = data[['Frequency','Reach']]
y = data[['Unique Clicks']]

lrModel.fit(x,y)

model = sm.OLS(y, x).fit()
predictions = model.predict(x)

model.summary()

```
